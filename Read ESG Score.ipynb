```python
def extract_details_from_filename(filename):
    parts = filename.replace('.csv', '').split('-')
    CompanyTicker = parts[0]
    Year = int(float(parts[1]))  # Convert to float first, then to int to handle any decimal points
    return CompanyTicker, Year
```

```python
# Set the path to the destination folder where the files will be copied
destination_folder = "Test/New Report Data"
NewDatafile = os.listdir(destination_folder)
# Loop through each file in the source folder
for filename in Datafile:
    CompanyTicker, Year = extract_details_from_filename(filename)
    if CompanyTicker in Matching_Ticker and Year >= 2022:
            # Copy the file to the destination directory
            source_path = os.path.join(DataPath, filename)
            destination_path = os.path.join(destination_folder, filename)
            shutil.copy(source_path, destination_path)
            print(f"Copied {filename} to {destination_folder}")

```
